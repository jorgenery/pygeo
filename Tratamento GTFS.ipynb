# Iniciando Carga de dados

```python
# Import package
import gtfs_functions as gtfs
import geopandas
import pygeos
import rtree

!curl -o GTFS_INTEGRA_SALVADOR.zip \
      -H 'Accept: application/vnd.github.v3.raw' \
      --location https://github.com/jorgenery/pygeo/blob/main/gtfs/GTFS_INTEGRA_SALVADOR.zip?raw=true
# Parse GTFS
routes, stops, stop_times, trips, shapes = gtfs.import_gtfs("GTFS_INTEGRA_SALVADOR.zip")
```

```python
## Mapeando Linhas
cutoffs = [0,6,9,15.5,19,22,24]
line_freq = gtfs.lines_freq(stop_times, trips, shapes, routes, cutoffs = cutoffs)

filtro_sentido = line_freq.dir_id == 'Inbound'
filtro_faixa = line_freq.window == '6:00-9:00'
filtro_linha = ["0720" in l for l in line_freq.route_name]

gdf = line_freq.loc[( filtro_sentido & filtro_faixa & filtro_linha ),:].reset_index()

gtfs.map_gdf(gdf = gdf, 
              variable = 'ntrips', 
              colors = ["#d13870", "#e895b3" ,'#55d992', '#3ab071', '#0e8955','#066a40'], 
              tooltip_var = ['route_name'] , 
              tooltip_labels = ['Linha: '], 
              breaks = [5, 10, 20, 50])
```

```python
# Grafico Frequencia
gtfs.map_gdf(gdf = gdf, 
             variable='ntrips',
             tooltip_var=['frequency'],
             tooltip_labels=['Frequency: '],
             breaks = [5, 10, 20, 50])
```
